# Visualization

```python
sns.barplot(data = Product_price.reset_index(), x = "Product Name", y = "mean")
plt.xticks.rotation = 90
plt.show()
```

```python
sns.barplot(data = df , x = "Product Name" , y = "Sales" , estimator="mean" , hue = "Category")
plt.xticks.rotation = 90
plt.show()
```

```python
sns.barplot(data = df , x = "Category" , y = "Quantity" , estimator="mean" , hue = "Sales")
plt.xticks.rotation = 90
plt.show()
```

```python
# plot barplot for the category and profit
plt.figure(figsize=(10,5))
sns.barplot(data = df , x = "Category" , y = "Profit" , estimator="mean")
plt.xticks(rotation = 90)
plt.title("Mean Profit for each Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
# # plot barplot for the Sales and country
plt.figure(figsize=(10,5))
sns.barplot(data = df , x = "Category" , y = "Sales" , estimator="mean" , hue = "Country")
plt.xticks(rotation = 90)
plt.title("Mean Sales for each Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
plt.scatter(df["Category"], df["Sales"], alpha=.3)
plt.title("Scatter plot of Sales vs Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
plt.scatter(df["Category"], df["Quantity"], alpha=.3)
plt.xticks(rotation=90)
plt.title("Scatter plot of Quantity vs Category")
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.show()
```

```python
categories = df["Category"].unique()
colors = plt.cm.rainbow(np.linspace(0, 1, len(categories)))

for category, color in zip(categories, colors):
	subset = df[df["Category"] == category]
	plt.scatter(subset["Product Name"], subset["Sales"], alpha=.3, label=category, color=color)

plt.legend()
plt.show()
```

```python
sns.pairplot(df, hue='Category')
```

```python
px.scatter_3d(data_frame = df , x = "Sales" , y = "Profit" , z = "Discount" , color = "Category")
```

```python
# plot the correlation matrix using seaborn
plt.figure(figsize=(10, 5))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

```python
# plot pie chart for the categories
plt.figure(figsize=(10, 5))
df['Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Categories Distribution')
plt.show()
```

```python
# plot pie chart for the markets
plt.figure(figsize=(10, 5))
df['Market'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Markets Distribution')
plt.show()
```

```python
# plot pie chart for the countries
plt.figure(figsize=(10, 5))
df['Country'].value_counts().head(10).plot.pie(autopct='%1.1f%%')
plt.title('Top 10 Countries Distribution')
plt.show()
```
